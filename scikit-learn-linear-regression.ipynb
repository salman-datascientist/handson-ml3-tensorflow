```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor   
```

```python
#Download and prepare the datas
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")

x = lifesat[["GDP per capita (USD)"]].values
y = lifesat["Life satisfaction"].values
```

```python
#Visualize the data
lifesat.plot(kind="scatter", grid=True, x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4,9])
plt.show()
```

```python
# Select a linear model
model = LinearRegression()
```

```python
# Train the model
model.fit(x, y)
```

```python
# Make a prediction for Cyprus
X_new = [[37_655.2]] # Cyprus' GDP per capita
print(model.predict(X_new)) # expected output: [[6.30165767 ]]
```

```python
# Select a K Neighbors Regressor model
model2 = KNeighborsRegressor(n_neighbors=3)
```

```python
# Train the model 2
model2.fit(x, y)
```

```python
# Make a prediction for Cyprus
X_new = [[37_655.2]] # Cyprus' GDP per capita
print(model2.predict(X_new)) # expected output: [[6.30165767 ]]
```
